```python
import pandas as pd
import numpy as np
PATH = "c:/school/SchoolProgram/NTUST_CSIE_DS/DataSet/" # 改成自己的路徑

preprocessing_basic = pd.read_csv(PATH + "preprocessing_basic.csv")
preprocessing_T1 = pd.read_csv(PATH + "preprocessing_T1.csv")
preprocessing_T2 = pd.read_csv(PATH + "preprocessing_T2.csv")
preprocessing_T3 = pd.read_csv(PATH + "preprocessing_T3.csv")

#merge attritube label
preprocessing_T1_basic = preprocessing_basic.merge(preprocessing_T1, on='acct')
preprocessing_T1_basic = preprocessing_T1_basic.loc[:,~preprocessing_T1_basic.columns.duplicated()] #drop deplicated columns
preprocessing_T1_basic.to_csv(PATH + "preprocessing_T1_basic.csv", index=False)

preprocessing_T1_2 = preprocessing_T1_basic.merge(preprocessing_T2, on='acct')
preprocessing_T1_2 = preprocessing_T1_2.loc[:,~preprocessing_T1_2.columns.duplicated()] #drop deplicated columns
preprocessing_T1_2.to_csv(PATH + "preprocessing_T1_2.csv", index=False)

preprocessing_T1_2_3 = preprocessing_T1_2.merge(preprocessing_T3, on='acct')
preprocessing_T1_2_3 = preprocessing_T1_2_3.loc[:,~preprocessing_T1_2_3.columns.duplicated()] #drop deplicated columns
preprocessing_T1_2_3.to_csv(PATH + "preprocessing_T1_2_3.csv", index=False)
```
